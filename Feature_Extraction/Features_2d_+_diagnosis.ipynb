## Putting All Together

### Join all dataframes from the features extraction notebooks in a single one called data_std_limit

```python
diagnosis_data = pd.read_csv('tcia-diagnosis-data-2012-04-20.csv')
diagnosis_data
```

#### std_limit = 0.5

```python
%cd "/home/cmonteiro/Aulas/Laboratório de IA e CD/Projeto 1/std_limit_0.5"
```

```python
features_radiomics_05 = pd.read_csv('features_radiomics_2d_0.5.csv')
features_xml_05 = pd.read_csv('features_pylidc_0.5.csv')
```

```python
data = pd.merge(features_xml_05, features_radiomics_05, on='Nodule_id', how='inner')
data_05 = data.merge(diagnosis_data, left_on=data.iloc[:, 1], right_on='TCIA Patient ID', how='left')
data_05.to_csv('data_0.5.csv', index=False)
data_05
```
