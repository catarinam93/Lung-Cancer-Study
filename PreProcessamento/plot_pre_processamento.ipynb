# Plot functions

```python
def plot_dataset(data, column_list=None, figsize=(8, 3)):

    # If column_list is not provided, select all numeric columns
    if column_list is None:
        numeric_columns = data.select_dtypes(include=['number']).columns.tolist()
    else:
        numeric_columns = column_list

    # Set the style for the plots (you can customize this)
    sns.set(style="whitegrid")
    
    # Create subplots for each numeric column
    num_plots = len(numeric_columns)
    fig, axes = plt.subplots(num_plots, 1, figsize=figsize)
    
    # Plot each numeric column
    for i, column in enumerate(numeric_columns):
        ax = axes[i]
        sns.histplot(data=data, x=column, kde=True, ax=ax)
        ax.set_title(f'Distribution of {column}')
        ax.set_xlabel(column)
    
    # Adjust layout and display the plots
    plt.tight_layout()
    plt.show()
```

## Radiologists

```python
plot_dataset(df_radiologists, figsize=(10, 20))
```

## Nodule

```python
plot_dataset(df_nodule, figsize=(10, 6))
```
