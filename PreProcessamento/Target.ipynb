# Criação da coluna target: média arredondada dos < malignancy >

```python
import os
import pylidc as pl
import pandas as pd
import shutil
```

```python
input_directory = "/home/cmonteiro/Aulas/Laboratório de IA e CD/Projeto 1/Imagens/LIDC-IDRI"
```

```python
# List that contains the malignancy of all patients
data = []

# Ordered list of all the subfolders
patient_folders = sorted(os.listdir(input_directory))

for patient_folder in patient_folders:
    patient_folder_path = os.path.join(input_directory, patient_folder)

    # Id of the patient - 'LIDC-IDRI-xxxx'
    patient_id = os.path.basename(patient_folder_path)
    
    patient_scans = pl.query(pl.Scan).filter(pl.Scan.patient_id == patient_id)
    
    # List that contains all the annotations about the malignancy of one patient
    malignancy_values = []

    for scan in patient_scans:
        anns = scan.annotations
        for ann in anns:
            malignancy_values.append(ann.malignancy)
    
    # Calculation of the round mean of the malignancy
    if malignancy_values:
        average_malignancy = round(sum(malignancy_values) / len(malignancy_values))
        
        # Dictionary with all the correspondancies
        malignancy_names = {
            1: "Highly Unlikely",
            2: "Moderately Unlikely",
            3: "Indeterminate",
            4: "Moderately Suspicious",
            5: "Highly Suspicious"
        }
        average_Malignancy = malignancy_names.get(average_malignancy)
        
        data.append([patient_id, f"{average_malignancy} - {average_Malignancy}"])
        
#supostamente não necessito disto pois já foram eliminadas as pastas no pre-processamento
#     else: 
#         print(patient_id) #tem que se eliminar estes pacientes porque não foram avaliados

#         # Caminho completo para a pasta a eliminar
#         delete_folder = os.path.join(input_directory, patient_id) 

#         # Verifique se a pasta existe
#         if os.path.exists(delete_folder):
#             # Verifique se é uma pasta (não um arquivo)
#             if os.path.isdir(delete_folder):
#                 # Exclua a pasta e seu conteúdo de forma recursiva
#                 shutil.rmtree(delete_folder)
#                 print(f"A pasta '{patient_id}' e seu conteúdo foram excluídos com sucesso.")
#             else:
#                 print(f"'{patient_id}' não é uma pasta.")
#         else:
#             print(f"A pasta '{patient_id}' não existe no caminho especificado.")

# Create a dataframe from the data list
target = pd.DataFrame(data, columns=["Patient_ID", "average-malignancy"])

# Set "Patient_ID" as the index of the dataframe
target.set_index("Patient_ID", inplace=True)

# Exportation of the DataFrame to a .csv file so it can be used on another notebook
target.to_csv('target.csv', index=False)

target
```
